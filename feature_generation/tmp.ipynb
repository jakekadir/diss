```python
recipes = pd.read_csv("../data/recipes.csv")
```

```python
df = pd.read_csv("trimmed_recipes_labelled_savoury.csv")

df["OriginalIndex"] = df["OriginalIndex"].astype("int64")

df = df.set_index("OriginalIndex")
```

```python
labelled_recipes = df.join(recipes[list(recipes.columns.difference(df.columns))], how="inner")
```

```python
labelled_recipes = labelled_recipes[["Savoury","Acidic","Name","RecipeIngredientParts","Description","RecipeId"]]
```

```python
labelled_recipes["RecipeIngredientParts"] = labelled_recipes["RecipeIngredientParts"].str[2:]
```

```python
labelled_recipes["RecipeIngredientParts"] = labelled_recipes["RecipeIngredientParts"].str[:-1]
```
