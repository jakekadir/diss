Import recommendation classes:

```python
from recipe2vec import Recipe2Vec
from sbert_embeddings import SBERTRecommender

DATASET_PATH = "../data/recipes.csv"
EXECUTION_ID = "trial_30_03"
```

Establish query to apply to all systems:

```python
ingredients = ["onion", "pasta", "beef"]
```

# SBERT Recommender

```python
bertender = SBERTRecommender(
    dataset_path=DATASET_PATH,
    col_to_embed="RecipeIngredientParts",
    execution_id=EXECUTION_ID,
)
```

```python
bertender.get_recommendations(ingredients, 10)
```

# Recipe2Vec

```python
recipe2vec = Recipe2Vec(
    dataset_path=DATASET_PATH, verbose=False, execution_id=EXECUTION_ID
)
```

```python
recipe2vec.get_recommendations(ingredients, 10)
```
